```python
import warnings
warnings.filterwarnings("ignore") #To avoid warnings.
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Top Indian Places to Visit.csv")
df
```

```python
print(df.isnull().sum())
```

```python
df.info()
```

```python
df = df[df['Entrance Fee in INR'] != 0]
df
```

```python
df = df.dropna()
```

```python
df.info()
```

```python
col = ['Type','Google review rating','Number of google review in lakhs','Entrance Fee in INR']
dfnew = df[col]
dfdummy = pd.get_dummies(dfnew,dtype=int)
```

```python
x = dfdummy.drop(columns = ['Entrance Fee in INR'])
y = dfdummy['Entrance Fee in INR']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30,shuffle = True)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x,y)
print(model.score(x_train,y_train))
print(model.score(x_test,y_test))
```

```python
from sklearn.linear_model import Lasso
reg1 = Lasso(alpha= 50,max_iter=100,tol= 0.1)
reg1.fit(x_train,y_train)
print(reg1.score(x_train,y_train))
print(reg1.score(x_test,y_test))
```

```python
from sklearn.linear_model import Ridge
reg2 = Ridge(alpha= 50,max_iter=100,tol= 0.1)
reg2.fit(x_train,y_train)
print(reg2.score(x_train,y_train))
print(reg2.score(x_test,y_test))
```
