```python
import warnings
warnings.filterwarnings("ignore") #To avoid warnings.
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("NY-House-Dataset.csv")
df
```

```python
df.info()
```

```python
df.nunique()
```

```python
col = ['TYPE','BEDS','BATH','STATE','ADMINISTRATIVE_AREA_LEVEL_2','LOCALITY','SUBLOCALITY','STREET_NAME','PRICE']
df1 = df[col]
newdf = pd.get_dummies(df1,dtype=int)
newdf
```

```python
x = newdf.drop(columns = ['PRICE'])
y = newdf['PRICE']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x,y)
print(model.score(x_train,y_train))
print(model.score(x_test,y_test))
```

```python
from sklearn.linear_model import Lasso
reg1 = Lasso(alpha= 25,max_iter=100,tol= 0.1)
reg1.fit(x_train,y_train)
print(reg1.score(x_train,y_train))
print(reg1.score(x_test,y_test))
```

```python
from sklearn.linear_model import Ridge
reg2 = Ridge(alpha= 25,max_iter=100,tol= 0.1)
reg2.fit(x_train,y_train)
print(reg2.score(x_train,y_train))
print(reg2.score(x_test,y_test))
```

```python
from sklearn.linear_model import LogisticRegression
loc = LogisticRegression()
loc.fit(x,y)
print(loc.score(x_train,y_train))
print(loc.score(x_test,y_test))
```
