```python
import warnings
warnings.filterwarnings("ignore") #To avoid warnings.
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("apple_quality.csv")
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
df['Quality'] = df['Quality'].replace('good',1)
df['Quality'] = df['Quality'].replace('bad',0)
df
```

```python
x = df.drop(columns = ['A_id'])
y = df['Quality']
```

```python
x.corr().style.background_gradient(cmap="Blues")
```

```python
x = x.drop(columns = ['Weight','Crunchiness','Ripeness','Acidity'])
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state= 2)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x,y)
print(model.score(x_train,y_train))
print(model.score(x_test,y_test))
```

```python
from sklearn.linear_model import Lasso
reg1 = Lasso(alpha= 25,max_iter=100,tol= 0.1)
reg1.fit(x_train,y_train)
print(reg1.score(x_train,y_train))
print(reg1.score(x_test,y_test))
```
